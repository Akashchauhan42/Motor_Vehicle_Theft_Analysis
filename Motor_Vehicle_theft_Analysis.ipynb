## Data cleaning

```python
# Drop the rows where non null values are less than 8.
df4=df4.dropna(thresh=8)
```

```python
# checking for null values
df4.isnull().sum()
```

```python
df4['vehicle_type'].unique()
```

```python
x=df4['vehicle_type'].mode()
x
```

```python
# Assigning missing values in the Vehicle_type column with the mode of vehicle_type
df4['vehicle_type']=df4['vehicle_type'].fillna('Stationwagon')
```

```python
df4.isnull().mean()*100
```

```python
# Assigning missing values in the Vehicle_desc column with the "unknown" word.
df4['vehicle_desc']=df4['vehicle_desc'].fillna('Unknown')
```

```python
df4.isnull().sum()
```

```python
# Drop the country column
df4.drop('country',axis=1,inplace=True)
```

```python
df4.shape
```

```python
# Converting the data type of Data_stolen column to 'datetime'.
df4['date_stolen']=pd.to_datetime(df4['date_stolen'])
```

```python
# Create a new column 'month' by accessing the month from date_stolen column
df4['month']=(df4['date_stolen']).dt.month
```

```python
#Create a new column 'year' by accessing the year from date_stolen column
df4['year']=(df4['date_stolen']).dt.year
```

```python
# Create a new column 'week_day' by accessing the day_name from date_stolen column
df4['week_day']=(df4['date_stolen']).dt.day_name()
```

```python
# Create a new column 'vehicle_age'
df4['vehicle_age']=2022-(df4['model_year'])
```

```python
# Create a new column 'vehicle_age_group'
df4['vehicle_age_group'] = pd.cut(
    df4['vehicle_age'],
    bins=[0, 15, 30, 45, 60, 75, 85],
    labels=['0-15', '15-30', '30-45', '45-60', '60-75','75-85'],
    include_lowest=True
)
```
